```python
class EstacionGasolina:
    def __init__(self, marca, numeroBombas, precioLitro, serviciosAdicionales):
        self.marca = marca
        self.numeroBombas = numeroBombas
        self.precioLitro = precioLitro
        self.serviciosAdicionales = serviciosAdicionales

    def surtirCombustible(self, litros):
        total = litros * self.precioLitro
        print("Surtiendo " + str(litros) + " litros en " + self.marca + ". Total: $" + str(total))

    def lavarAuto(self):
        print("El auto esta siendo lavado en la estacion " + self.marca)

    def revisarNiveles(self):
        print("Revisando niveles de aceite y agua en " + self.marca)
#Gasolinera 1
print("Configuracion de la Estacion de Gasolina 1")
nombre_marca1 = input("Introduce la marca de la primera estacion: ")
bombas1 = int(input("Introduce el numero de bombas: "))
precio1 = float(input("Introduce el precio por litro: "))
servicios1 = input("Introduce los servicios adicionales: ")

gasolinera1 = EstacionGasolina(nombre_marca1, bombas1, precio1, servicios1)

#Gasolinera 2
print("\nConfiguracion de la Estacion de Gasolina 2")
nombre_marca2 = input("Introduce la marca de la segunda estacion: ")
bombas2 = int(input("Introduce el numero de bombas: "))
precio2 = float(input("Introduce el precio por litro: "))
servicios2 = input("Introduce los servicios adicionales: ")

gasolinera2 = EstacionGasolina(nombre_marca2, bombas2, precio2, servicios2)

# Interaccion gasolinera1
print("\n--- Operando Gasolinera 1: " + gasolinera1.marca + " ---")
litros_cargar1 = float(input("Cuantos litros quieres cargar?: "))
gasolinera1.surtirCombustible(litros_cargar1)
gasolinera1.lavarAuto()

# Interaccion gasolinera2
print("\n--- Operando Gasolinera 2: " + gasolinera2.marca + " ---")
litros_cargar2 = float(input("Cuantos litros quieres cargar?: "))
gasolinera2.surtirCombustible(litros_cargar2)
gasolinera2.revisarNiveles()

print("\nPrograma realizado por Zavala Medina Alexia Paola NL= 54")
```
